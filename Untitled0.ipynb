````python
from sentence_transformers import SentenceTransformer

# Sentence Transformer 모델 초기화 (한국어 임베딩 생성 가능한 어떤 모델도 가능)
model = SentenceTransformer("snunlp/KR-SBERT-V40K-klueNLI-augSTS")


# SetntenceTransformer를 이용하여 임베딩 생성
def get_embedding(sentences):
    return model.encode(sentences)


# 주어진 문서의 리스트에서 배치 단위로 임베딩 생성
def get_embeddings_in_batches(docs, batch_size=100):
    batch_embeddings = []
    for i in range(0, len(docs), batch_size):
        batch = docs[i:i + batch_size]
        contents = [doc["content"] for doc in batch]
        embeddings = get_embedding(contents)
        batch_embeddings.extend(embeddings)
        print(f'batch {i}')
    return batch_embeddings


# 새로운 index 생성
def create_es_index(index, settings, mappings):
    # 인덱스가 이미 존재하는지 확인
    if es.indices.exists(index=index):
        # 인덱스가 이미 존재하면 설정을 새로운 것으로 갱신하기 위해 삭제
        es.indices.delete(index=index)
    # 지정된 설정으로 새로운 인덱스 생성
    es.indices.create(index=index, settings=settings, mappings=mappings)


# 지정된 인덱스 삭제
def delete_es_index(index):
    es.indices.delete(index=index)


# Elasticsearch 헬퍼 함수를 사용하여 대량 인덱싱 수행
def bulk_add(index, docs):
    # 대량 인덱싱 작업을 준비
    actions = [
        {
            '_index': index,
            '_source': doc
        }
        for doc in docs
    ]
    return helpers.bulk(es, actions)


# 역색인을 이용한 검색
def sparse_retrieve(query_str, size):
    query = {
        "match": {
            "content": {
                "query": query_str
            }
        }
    }
    return es.search(index="test", query=query, size=size, sort="_score")


# Vector 유사도를 이용한 검색
def dense_retrieve(query_str, size):
    # 벡터 유사도 검색에 사용할 쿼리 임베딩 가져오기
    query_embedding = get_embedding([query_str])[0]

    # KNN을 사용한 벡터 유사성 검색을 위한 매개변수 설정
    knn = {
        "field": "embeddings",
        "query_vector": query_embedding.tolist(),
        "k": size,
        "num_candidates": 100
    }

    # 지정된 인덱스에서 벡터 유사도 검색 수행
    return es.search(index="test", knn=knn)
````

````python
# 색인을 위한 setting 설정
settings = {
    "analysis": {
        "analyzer": {
            "nori": {
                "type": "custom",
                "tokenizer": "nori_tokenizer",
                "decompound_mode": "mixed",
                "filter": ["nori_posfilter"]
            }
        },
        "filter": {
            "nori_posfilter": {
                "type": "nori_part_of_speech",
                # 어미, 조사, 구분자, 줄임표, 지정사, 보조 용언 등
                "stoptags": ["E", "J", "SC", "SE", "SF", "VCN", "VCP", "VX"]
            }
        }
    }
}

# 색인을 위한 mapping 설정 (역색인 필드, 임베딩 필드 모두 설정)
mappings = {
    "properties": {
        "content": {"type": "text", "analyzer": "nori"},
        "embeddings": {
            "type": "dense_vector",
            "dims": 768,
            "index": True,
            "similarity": "l2_norm"
        }
    }
}
````

````python
# 필요한 라이브러리 임포트
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import torch
import traceback

# BitsAndBytesConfig 설정
bnb_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.bfloat16
)

# 모델 로드 (모델 이름을 실제 "야놀자" 모델로 변경해야 합니다)
model_name = "yanolja/EEVE-Korean-Instruct-10.8B-v1.0"  # 실제 모델 경로로 변경 필요

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    device_map='auto',
    quantization_config=bnb_config,
    torch_dtype=torch.float16
)
````

````python
# RAG 구현에 필요한 Question Answering을 위한 LLM 프롬프트
persona_qa = """
당신은 과학 상식 전문가입니다.

- 사용자의 이전 메시지 정보 및 주어진 Reference 정보를 활용하여 간결하게 답변을 생성합니다.
- 주어진 검색 결과 정보로 대답할 수 없는 경우는 정보가 부족해서 답을 할 수 없다고 대답합니다.
- 한국어로 답변을 생성합니다.
"""

# RAG 구현에 필요한 질의 분석 및 검색 이외의 일반 질의 대응을 위한 LLM 프롬프트
persona_function_calling = """
당신은 과학 상식 전문가입니다.

- 사용자가 대화를 통해 과학 지식에 관한 주제로 질문하면 standalone_query를 생성하여 검색을 수행해야 합니다.
- 사용자의 메시지에서 standalone_query를 추출하여 JSON 형식으로 출력하십시오.
- 만약 검색이 필요하지 않다면, standalone_query를 빈 문자열로 출력하십시오.

형식:
{
    "standalone_query": "<standalone_query>"
}
"""
````

````python
def answer_question(messages):
    response = {"standalone_query": "", "topk": [], "references": [], "answer": ""}

    # Standalone Query 생성
    prompt = persona_function_calling + "\n\n"
    for msg in messages:
        if msg["role"] == "user":
            prompt += "사용자: " + msg["content"] + "\n"
        elif msg["role"] == "assistant":
            prompt += "조수: " + msg["content"] + "\n"
    prompt += "JSON 형식으로 standalone_query를 출력하십시오.\n"

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(**inputs, max_new_tokens=256)
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # Standalone Query 파싱
    try:
        result_json = json.loads(generated_text)
        standalone_query = result_json.get("standalone_query", "")
    except json.JSONDecodeError:
        standalone_query = ""

    if standalone_query:
        # 검색 수행
        search_result = sparse_retrieve(standalone_query, 3)

        response["standalone_query"] = standalone_query
        retrieved_context = []
        for i, rst in enumerate(search_result['hits']['hits']):
            retrieved_context.append(rst["_source"]["content"])
            response["topk"].append(rst["_source"]["docid"])
            response["references"].append({"score": rst["_score"], "content": rst["_source"]["content"]})

        # 답변 생성
        answer_prompt = persona_qa + "\n\n"
        for msg in messages:
            if msg["role"] == "user":
                answer_prompt += "사용자: " + msg["content"] + "\n"
            elif msg["role"] == "assistant":
                answer_prompt += "조수: " + msg["content"] + "\n"
        answer_prompt += "다음은 검색 결과입니다:\n"
        for idx, context in enumerate(retrieved_context):
            answer_prompt += f"[{idx+1}] {context}\n"
        answer_prompt += "위의 정보를 참고하여 답변을 생성하십시오.\n"

        inputs = tokenizer(answer_prompt, return_tensors="pt").to(model.device)
        outputs = model.generate(**inputs, max_new_tokens=256)
        answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
        response["answer"] = answer

    else:
        # 검색이 필요 없는 경우 직접 답변 생성
        answer_prompt = persona_function_calling + "\n\n"
        for msg in messages:
            if msg["role"] == "user":
                answer_prompt += "사용자: " + msg["content"] + "\n"
            elif msg["role"] == "assistant":
                answer_prompt += "조수: " + msg["content"] + "\n"
        answer_prompt += "적절한 대답을 생성하십시오.\n"

        inputs = tokenizer(answer_prompt, return_tensors="pt").to(model.device)
        outputs = model.generate(**inputs, max_new_tokens=256)
        answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
        response["answer"] = answer

    return response
````

````python
def answer_question(messages):
    response = {"standalone_query": "", "topk": [], "references": [], "answer": ""}

    # Standalone Query 생성
    prompt = persona_function_calling + "\n\n대화:\n"
    for msg in messages:
        if msg["role"] == "user":
            prompt += "사용자: " + msg["content"] + "\n"
        elif msg["role"] == "assistant":
            prompt += "조수: " + msg["content"] + "\n"
    prompt += "\n위 사용자의 메시지를 기반으로 standalone_query를 생성하고, 아래 형식으로 출력하십시오:\n"
    prompt += '```json\n{"standalone_query": "<standalone_query>"}\n```\n'
    prompt += "그 외의 다른 출력은 하지 마십시오."

    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(
        **inputs,
        max_new_tokens=256,
        do_sample=False,
        temperature=0,
        eos_token_id=tokenizer.encode('```', add_special_tokens=False)[0]
    )
    generated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)

    # '```json' 이후부터 '```' 이전까지 추출
    import re
    pattern = r'```json\n(.*?)\n```'
    match = re.search(pattern, generated_text, re.DOTALL)
    if match:
        json_str = match.group(1)
        try:
            result_json = json.loads(json_str)
            standalone_query = result_json.get("standalone_query", "")
            response["standalone_query"] = standalone_query
        except json.JSONDecodeError:
            standalone_query = ""
    else:
        standalone_query = ""

    if standalone_query:
        # 검색 수행
        search_result = dense_retrieve(standalone_query, 3)

        retrieved_context = []
        for i, rst in enumerate(search_result['hits']['hits']):
            retrieved_context.append(rst["_source"]["content"])
            response["topk"].append(rst["_source"]["docid"])
            response["references"].append({"score": rst["_score"], "content": rst["_source"]["content"]})

        # 답변 생성
        answer_prompt = persona_qa + "\n\n"
        for msg in messages:
            if msg["role"] == "user":
                answer_prompt += "사용자: " + msg["content"] + "\n"
            elif msg["role"] == "assistant":
                answer_prompt += "조수: " + msg["content"] + "\n"
        answer_prompt += "\n다음은 검색 결과입니다:\n"
        for idx, context in enumerate(retrieved_context):
            answer_prompt += f"[{idx+1}] {context}\n"
        answer_prompt += "\n위의 정보를 참고하여 사용자에게 답변을 생성하십시오."

        inputs = tokenizer(answer_prompt, return_tensors="pt").to(model.device)
        outputs = model.generate(
            **inputs,
            max_new_tokens=256,
            do_sample=True,
            temperature=0.7,
            top_p=0.9,
            eos_token_id=tokenizer.eos_token_id
        )
        answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
        # 프롬프트 부분 제거
        answer = answer[len(answer_prompt):].strip()
        response["answer"] = answer

    else:
        # 검색이 필요 없는 경우 직접 답변 생성
        answer_prompt = persona_function_calling + "\n\n"
        for msg in messages:
            if msg["role"] == "user":
                answer_prompt += "사용자: " + msg["content"] + "\n"
            elif msg["role"] == "assistant":
                answer_prompt += "조수: " + msg["content"] + "\n"
        answer_prompt += "\n적절한 대답을 생성하십시오."

        inputs = tokenizer(answer_prompt, return_tensors="pt").to(model.device)
        outputs = model.generate(
            **inputs,
            max_new_tokens=256,
            do_sample=True,
            temperature=0.7,
            top_p=0.9,
            eos_token_id=tokenizer.eos_token_id
        )
        answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
        # 프롬프트 부분 제거
        answer = answer[len(answer_prompt):].strip()
        response["answer"] = answer

    return response
````

````python
def eval_rag(eval_filename, output_filename):
    with open(eval_filename) as f, open(output_filename, "w") as of:
        idx = 0
        for line in f:
            if idx > 5:
                break
            j = json.loads(line)
            print(f'Test {idx}\nQuestion: {j["msg"]}')
            response = answer_question(j["msg"])
            print(f'Answer: {response["answer"]}\n')

            # 대회 score 계산은 topk 정보를 사용, answer 정보는 LLM을 통한 자동평가시 활용
            output = {"eval_id": j["eval_id"], "standalone_query": response["standalone_query"], "topk": response["topk"], "answer": response["answer"], "references": response["references"]}
            of.write(f'{json.dumps(output, ensure_ascii=False)}\n')
            idx += 1
````

````python
def answer_question_without_rag(messages):
    # 프롬프트 설정
    prompt = """
당신은 과학 상식 전문가입니다.

- 사용자의 질문에 대해 정확하고 간결하게 답변을 제공합니다.
- 한국어로 답변을 생성합니다.

대화:
"""
    for msg in messages:
        if msg["role"] == "user":
            prompt += "사용자: " + msg["content"] + "\n"
        elif msg["role"] == "assistant":
            prompt += "조수: " + msg["content"] + "\n"
    prompt += "조수:"

    # 모델에게 프롬프트 전달하여 답변 생성
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    outputs = model.generate(
        **inputs,
        max_new_tokens=256,
        do_sample=True,
        temperature=0.7,
        top_p=0.9
    )
    answer = tokenizer.decode(outputs[0], skip_special_tokens=True)
    # 프롬프트 부분은 제거하고, 모델의 응답만 추출
    answer = answer[len(prompt):].strip()

    return {"answer": answer}
````

````python
def eval_without_rag(eval_filename, output_filename):
    with open(eval_filename, encoding='utf-8') as f, open(output_filename, "w", encoding='utf-8') as of:
        idx = 0
        for line in f:
            if idx > 5:
                break
            j = json.loads(line)
            print(f'Test {idx}\nQuestion: {j["msg"][0]["content"]}')
            response = answer_question_without_rag(j["msg"])
            print(f'Answer: {response["answer"]}\n')

            # 결과를 저장 (topk와 references는 빈 리스트로 저장)
            output = {
                "eval_id": j["eval_id"],
                "answer": response["answer"],
                "topk": [],
                "references": []
            }
            of.write(f'{json.dumps(output, ensure_ascii=False)}\n')
            idx += 1
````
