```python
import pandas as pd
import json

# eval.jsonl 파일에서 데이터 읽기
rows = []
with open('/ir/data/eval.jsonl', 'r', encoding='utf-8') as f:
    for line in f:
        # 각 줄을 개별적으로 파싱
        item = json.loads(line.strip())
        eval_id = item['eval_id']
        messages = item['msg']
        
        # 여러 role과 content를 각각의 role_1, content_1 형태로 변환
        row = {"eval_id": eval_id}
        for i, message in enumerate(messages, 1):
            row[f'role_{i}'] = message['role']
            row[f'content_{i}'] = message['content']
        
        rows.append(row)

# DataFrame으로 변환
df = pd.DataFrame(rows)

# 결과 출력 또는 저장
print(df)
# df.to_csv('output.csv', index=False)  # CSV로 저장하고 싶다면

```

```python
import pandas as pd
import json

# JSONL 파일 경로
jsonl_file_path = '/ir/data/documents.jsonl'

# JSONL 파일을 읽어서 pandas 데이터프레임으로 변환
def read_jsonl_file(file_path):
    # 한 줄씩 읽어와서 json 데이터로 변환
    data = [json.loads(line) for line in open(file_path, 'r', encoding='utf-8')]
    
    # 데이터프레임으로 변환
    df = pd.DataFrame(data)
    
    return df

# 파일 읽기 및 변환
df = read_jsonl_file(jsonl_file_path)

# 변환된 데이터프레임 출력
print(df.head())  # 첫 몇 줄만 출력
```

```python
df['content_length'] = df['content'].apply(len)

# 최고, 최저, 평균 길이 계산
max_length = df['content_length'].max()
min_length = df['content_length'].min()
mean_length = df['content_length'].mean()

print(f"최고 길이: {max_length}")
print(f"최저 길이: {min_length}")
print(f"평균 길이: {mean_length}")
```

```python
df['word_count'] = df['content'].apply(lambda x: len(x.split()))

# 최고, 최저, 평균 단어 수 계산
max_word_count = df['word_count'].max()
min_word_count = df['word_count'].min()
mean_word_count = df['word_count'].mean()

print(f"최고 단어 수: {max_word_count}")
print(f"최저 단어 수: {min_word_count}")
print(f"평균 단어 수: {mean_word_count}")
```
